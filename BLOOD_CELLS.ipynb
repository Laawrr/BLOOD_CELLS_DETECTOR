# IMPORTS

```python
import os
import yaml
from ultralytics import YOLO
import torch
from IPython.display import Image
```

# CONFIGURATION

```python
DATASET_DIR = os.path.abspath("./datasets/blood_cells") 
DATA_YAML_PATH = os.path.join(DATASET_DIR, "data.yaml")
PROJECT_NAME = "blood_cell_project"
RUN_NAME = "yolov8_blood_cells_v1"
EPOCHS = 50                 
IMG_SIZE = 640             
BATCH_SIZE = 16             
MODEL_TYPE = "yolov8s.pt" 
```

# CHECK ENVIRONMENT

```python
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Device: {torch.cuda.get_device_name(0)}")
else:
    print("WARNING: Running on CPU. Training will be slow.")
```

# VERIFY DATA.YML

```python
if os.path.exists(DATA_YAML_PATH):
    with open(DATA_YAML_PATH, 'r') as f:
        data_config = yaml.safe_load(f)

    data_config['path'] = DATASET_DIR
else:
    print(f"Error: Could not find {DATA_YAML_PATH}. Please check your path.")
```

# LOAD MODEL

```python
print(f"Loading {MODEL_TYPE}...")
model = YOLO(MODEL_TYPE)
```

# TRAIN

```python
print("Starting training...")
results = model.train(
    data=DATA_YAML_PATH,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    project=PROJECT_NAME,
    name=RUN_NAME,
    patience=10,        
    verbose=True
)
```

# DISPLAY IMPORTANT METRICS

```python
print("\nAll available metrics:")
for key, value in results.results_dict.items():
    print(f"{key}: {value:.4f}")
```

# DISPLAY TRAINING GRAPHS

```python
base_path = os.path.join(PROJECT_NAME, RUN_NAME)
print (base_path)
print("\nTraining Curves:")
if os.path.exists(os.path.join(base_path, "results.png")):
    display(Image(filename=os.path.join(base_path, "results.png")))

print("\nPrecision-Recall Curve:")
if os.path.exists(os.path.join(base_path, "PR_curve.png")):
    display(Image(filename=os.path.join(base_path, "PR_curve.png")))

print("\nConfusion Matrix:")
if os.path.exists(os.path.join(base_path, "confusion_matrix.png")):
    display(Image(filename=os.path.join(base_path, "confusion_matrix.png")))
```

# EXPORT MODEL

```python
export_formats = ["onnx", "torchscript"] 

for fmt in export_formats:
    print(f"Exporting model to {fmt} format...")
    model.export(format=fmt)
```
